```python
from sympy import *
```

```python
a = 0.6
b = 5.5
c = 10
d = -5
```

```python
x = Symbol("x")
```

```python
Fx = a*(x**3) + b*(x**2) + c*x + d
```

```python
Fx
```

```python
solve(Fx, x)
```

Определим корни уравнения:

```python
n = 1
for el in solve(Fx, x):
  print (f'корень № {n}: {el}')
  n+=1
```

Построим график функции:

```python
from sympy.plotting import plot
```

```python
graph = plot(Fx, (x, -15, 10), line_color = 'green', title = 'график функции',legend = 'True')
```

```python
dif = diff(Fx)
```

```python
dif
```

```python
graph_dif = plot(dif, (x, -15, 10), line_color = 'red', title = 'график производной функции',legend = 'True')
```

```python
graph_result = plot(Fx, dif, (x, -6, 3),title = 'график функции и ее производной', legend = 'True', xlabel = 'X', ylabel = "Y", show = False)
graph_result[0].line_color = 'green'
graph_result[1].line_color = 'red'
graph_result.show()
```

найдем экстремумы, приравняв производную к нулю (эти же точки являются вершинами крафика исходной функции):

```python
n = 1
for el in solve(dif, x):
  print (f'вершина № {n}: x = {el}')
  n+=1
```

интервалы возрастания функции (на которых производная больше нуля):

```python
solve(dif > 0, x)
```

интервалы убывания функции (на которых производная меньше нуля):

```python
solve(dif < 0, x)
```

интервалы, на которых функция Fx больше нуля:

```python
solve(Fx > 0, x)
```

интервалы, на которых функция Fx меньше нуля:

```python
solve(Fx < 0, x)
```
